```python
#create tuples
channels = cv2.split(image)
#create color tuples
colors = ('blue','green','red')
#create plot
plt.figure()
plt.title('color histograms')
plt.xlabel('pixel value')
plt.ylabel('#of pixels')


i=0
for channel in channels:
    hist = cv2.calcHist([channel], [0], None, [255], [0, 256])
    plt.plot(hist, color=colors[i])
    i+=1

plt.show()
```

```python
filepath1 = os.path.join('..',
                        '..',
                        '..',
                        'data',
                        'img',
                        'trex.png')

filepath2 = os.path.join('..',
                        '..',
                        '..',
                        'data',
                        'img',
                        't-rex2.png')
```

```python
image1=cv2.imread(filepath1)

image2=cv2.imread(filepath2)
```

```python
hist1 = cv2.calcHist([image1], [0,1,2], None, [255,255,255], [0,256, 0,256, 0,256])
hist2 = cv2.calcHist([image2], [0,1,2], None, [255,255,255], [0,256, 0,256, 0,256])
```

```python
normalizedHist1 = cv2.normalize(hist1, hist1, 0, 1.0, cv2.NORM_MINMAX)
normalizedHist2 = cv2.normalize(hist2, hist2, 0, 1.0, cv2.NORM_MINMAX)
```

```python
round(cv2.compareHist(normalizedHist1, normalizedHist2, cv2.HISTCMP_CHISQR),2)
```
